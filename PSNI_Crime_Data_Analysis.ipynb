# What the headings mean
## Reported by: 
The force that provided the data about the crime.

## Falls within: 
At present, also the force that provided the data about the crime. This is currently being looked into and is likely to change in the near future.

## Longitude and Latitude:
The anonymised coordinates of the crime. See Location Anonymisation for more information.

The latitude and longitude locations of Crime and ASB incidents published on this site always represent the approximate location of a crime â€” not the exact place that it happened.

When crime data is uploaded by police forces, the exact location of each crime is compared against this master list to find the nearest map point. The co-ordinates of the actual crime are then replaced with the co-ordinates of the map point. If the nearest map point is more than 20km away, the co-ordinates are zeroed out. No other filtering or rules are applied.

## LSOA code and LSOA name:
References to the Lower Layer Super Output Area that the anonymised point falls into, according to the LSOA boundaries provided by the Office for National Statistics.

## Crime type:
One of the crime types listed in the Police.UK FAQ.

Some categories can be quite vague and include types of crime that vary quite a but. e.g. Violence and sexual offences groups together common assaults, Grievous Bodily Harm and sexual offences. It could be useful to have the numbers for each of these separately.

## Last outcome: category:
A reference to whichever of the outcomes associated with the crime occurred most recently. For example, this crime's 'Last outcome category' would be 'Formal action is not in the public interest'.

## Context:
A field provided for forces to provide additional human-readable data about individual crimes. Currently, for newly added CSVs, this is always empty.

```python
crimesSample = crimes.sample(10000)
crimesSample = crimesSample.reset_index(drop=True)
print(crimesSample.head())
```

I decided to take a sample of crimes to geocode with as it would take far too long to process all crimes in the dataset.

```python
crimesSample.head()
```

This was just to check that my new sample data was still similar to the original dataframe.

```python
import geocoder

crimesSample['geocoded json'] = ""

pd.options.mode.chained_assignment = None

def getAddressesFromTable(table):
    count = 0
    url = 'http://makerspace.eeecs.qub.ac.uk/nominatim/'   
    for index, row in table.iterrows():
        try:
            g = geocoder.osm([row['Latitude'],
                              row['Longitude']],
                             method='reverse', url=url)
            if g is not None:
                json = g.json
                table['geocoded json'][index] = json
                
        except:
            #Don't lose what has already been processed if failure occurs
            return table
    return table

crimesSampleWithGeoLoc = getAddressesFromTable(crimesSample)
crimesSampleWithGeoLoc.to_csv('data/crimes_sample.csv',
                              encoding='utf-8')

print('Data frame saved as .csv file')
```

I used geocoding on the sample data to get location information using the coordinates. I then saved the resulting data frame into a csv file so that in the future I won't have to wait on the geocoding.

# Processing results of geocoding

```python
import pandas as pd
import ast

crimes = pd.read_csv("data/crimes_sample.csv",index_col = 0)

def fixJsonString(jsonString):
    try:
        return ast.literal_eval(jsonString)
    except Exception as e:
        print(e)
        print(jsonString)


crimes["geocoded json"] = crimes["geocoded json"].apply(fixJsonString)
crimes.head()
```

I processed the results of the geocoding so that it could be more easily used by python. Above is the resulting dataframe after the geocoding and processing.

```python
pd.options.mode.chained_assignment = None
crimes["Address"] = ""
crimes["Town"] = ""
crimes["County"] = ""
crimes["Street"] = ""

for i in range(0, len(crimes)):
    crimes["Address"][i] = crimes["geocoded json"][i].get("address")
    if 'city' in crimes["geocoded json"][i]:
        crimes["Town"][i] = crimes["geocoded json"][i].get("city")
    elif 'town' in crimes["geocoded json"][i]:
        crimes["Town"][i] = crimes["geocoded json"][i].get("town")
    elif 'village' in crimes["geocoded json"][i]:
        crimes["Town"][i] = crimes["geocoded json"][i].get("village")
    else:
        crimes["Town"][i] = "None"
    crimes["County"][i] = crimes["geocoded json"][i].get("county")
    crimes["Street"][i] = crimes["geocoded json"][i].get("street")
    
crimes = crimes.drop(columns=['Location'])

crimes.head()
```

```python
bars = pd.read_csv("data/Pubs_and_Clubs_in_Causeway_Coast_and_Glens.csv",
                   index_col = 0)
bars.head()
```

I wanted to see if the number of bars in a town has an effect on it's crime. I was able to find a dataset which contains the names and locations of bars in towns on the causeway coast.

```python
import numpy as np

barTowns = pd.DataFrame(bars['Town'].value_counts())
barTowns = barTowns.reset_index(drop=False)
barTowns.columns = ['Town', 'Bars']
barTowns = barTowns.replace(' ', np.nan).dropna()
barTowns.head()
```

Beacuse I only needed the numner of bars in each town, I cleaned the bar data to just show towns and the number of bars that they have.

```python
def addBars(townName):
    try:
        return barTowns.loc[(barTowns["Town"] == townName),
                            "Bars"].values[0]
    except:
        return np.NaN

townCrimeCounts["Bars"] = townCrimeCounts["Town"].apply(addBars)
townCrimeCounts = townCrimeCounts.dropna()
townCrimeCounts = townCrimeCounts.reset_index(drop=True)
townCrimeCounts.head()
```

I combined the bar data with the crimes data. Now I can see the numbers of crimes in a town alongside the number of bars.

```python
populations = pd.read_csv("data/cc_populations.csv", index_col = 0)
populations
```

In order to get a better idea on if the numbers of bars in a town has an effect on crime, I manually created a csv file which contains the populations of the causeaway cost towns from wikipedia. I'm going to use this to normalise the crime and bar count data.

```python
townCrimeCounts["Population"] = populations["Population"]
townCrimeCounts.head()
```

```python
def townCrimeCountsPerTenThousand(row):
    population = row["Population"]
    ratio = population/10000
    for i in range( 1, 16):
        row[i] = row[i] / ratio
    return row

townCrimeCountsPerTenThousand = townCrimeCounts.apply(townCrimeCountsPerTenThousand,
                                                      axis=1)

townCrimeCountsPerTenThousand = townCrimeCountsPerTenThousand\
.rename(columns={'Anti-social behaviour': 'Anti-social behaviour per 10,000',
                 'Other crime': 'Other crime per 10,000',
                 'Public order': 'Public order per 10,000',
                 'Vehicle crime': 'Vehicle crime per 10,000',
                 'Burglary': 'Burglary per 10,000',
                 'Robbery': 'Robbery per 10,000',
                 'Other theft': 'Other theft per 10,000',
                 'Violence and sexual offences': 'Violence and sexual offences per 10,000',
                 'Criminal damage and arson': 'Criminal damage and arson per 10,000',
                 'Drugs': 'Drugs per 10,000',
                 'Possession of weapons': 'Possession of weapons per 10,000',
                 'Shoplifting': 'Shoplifting per 10,000',
                 'Theft from the person': 'Theft from the person per 10,000',
                 'Bicycle theft': 'Bicycle theft per 10,000',
                 'Bars': 'Bars per 10,000'})

townCrimeCountsPerTenThousand.head()
```

```python
barTownsSplitBars = pd.DataFrame(bars[["Town","Type_"]])
barTownsSplitBars = barTownsSplitBars.reset_index(drop=True)
barTownsSplitBars.columns = ['Town', 'Bar Type']
barTownsSplitBars = barTownsSplitBars.groupby(["Town", "Bar Type"])\
.size().reset_index(name="Count")
barTownsSplitBars = barTownsSplitBars.replace(' ', np.nan).dropna()
barTownsSplitBars = barTownsSplitBars.sort_values(['Town',"Bar Type"],
                                                  ascending=True)

#Building a crime dictionary
townsSplitBars = barTownsSplitBars["Town"].drop_duplicates()\
.reset_index(drop=True)
barTypesSplitBars = barTownsSplitBars["Bar Type"].drop_duplicates()\
.reset_index(drop=True)


barTypeDict = {}
#Generate the crime counts
for i,town in townsSplitBars.iteritems():
    barTypeList = []
    for i,barType in barTypesSplitBars.iteritems():
        try:
            barCount = barTownsSplitBars\
            .loc[(barTownsSplitBars["Town"] == town) &\
                 (barTownsSplitBars["Bar Type"] == barType),
                 "Count"].values[0]
        except:
            barCount = 0
        barTypeList.append(barCount)
    barTypeDict.update({town : barTypeList })

    

columnHeadings = ["Town"]
for i in range(0, len(barTypesSplitBars)):
    columnHeadings.append(barTypesSplitBars[i])

barTypesSplitBars = pd.DataFrame(columns=columnHeadings)

for town in townsSplitBars:
    barTypesSplitBars = barTypesSplitBars\
    .append({'Town':town,
             'Pub':barTypeDict.get(town)[0],
             'Restaurant':barTypeDict.get(town)[1],
             'Hotel':barTypeDict.get(town)[2],
             'Night Club':barTypeDict.get(town)[3],
             'Social Club':barTypeDict.get(town)[4]},
            ignore_index=True)
```

# Google Places API

In this section I used Google's places API in order to get the coordinated of nearby bars to crimes.

I decided that I would re run the google places tests on the original dataset. This will allow me to use a larger sample which will remove the low numbers od certain crime types.

```python
import pandas as pd
crimes = pd.read_csv('data/all_crimes.csv',
                     encoding='utf-8').drop(columns=['Unnamed: 0']).reset_index(drop=True)
crimes.dropna(subset=['Longitude'], inplace=True)
print(crimes["Longitude"].isna().sum())
print(crimes["Latitude"].isna().sum())
crimes = crimes.drop(columns=['Month','Location','Crime ID','LSOA code',
                              'LSOA name','Last outcome category',
                              'Context','Reported by', 'Falls within'])
crimes.info()
crimes.head()
```

I loaded in a data set saved at the beginning of my notebook which contains all crime information. I cleaned it to only contain necessary data.

```python
crimes['Crime type'].value_counts()
```

```python
def reduceType(df, crimeType):
    crimeWithType = df[df["Crime type"] == crimeType]
    if len(crimeWithType) >= 150:
        crimeWithTypeSample = crimeWithType.sample(150)
    else:
        crimeWithTypeSample = crimeWithType
    return crimeWithTypeSample.reset_index()

crimeTypes = set(crimes["Crime type"].tolist())

typeDfs = []

for crimeType in crimeTypes:
    typeDfs.append(reduceType(crimes, crimeType))
        
count = 0
for i in range(0, len(typeDfs)):
    typeDfs[i] = typeDfs[i].drop(columns=['index'])
    count += len(typeDfs[i])
    
print("%d total crimes" % (count))

combinedTypeDfs = pd.concat(typeDfs, ignore_index=True)
```

Beacuse of the daily use limit of google places api, as well as time constraints, I had to reduce the number of values I processed. I created a sample which contains 150 crimes of each crime type.

```python
from googleplaces import GooglePlaces, types, lang

#I have removed my api key from the code for security
API_KEY = 'my_api_key'
google_places = GooglePlaces(API_KEY)

def getClosestBarGeoLoc(longitude, latitude):  
    query_result = google_places.nearby_search(
        lat_lng={'lng': longitude, 'lat': latitude},radius=50000,
        types=[types.TYPE_BAR],rankby='distance')
    return(query_result.places[0].geo_location)
```

```python
combinedTypeDfs["Longitude of closest bar"] = ''
combinedTypeDfs["Latitude of closest bar"] = ''

for i in range(0, len(combinedTypeDfs)):
        try:
            longitude = combinedTypeDfs["Longitude"][i]
            latitude = combinedTypeDfs["Latitude"][i]
            geoLoc = getClosestBarGeoLoc(longitude, latitude)
            combinedTypeDfs["Longitude of closest bar"][i] = geoLoc.get('lng') 
            combinedTypeDfs["Latitude of closest bar"][i] = geoLoc.get('lat')
        except:
            pass
```
